```python
import pandas as pd
import json
```

```python
df = pd.read_csv('reviews.csv')
df.head()
```

```python
df.nunique()
```

```python
# check which cols to drop

cols = [
    'brand',
    'colors',
    'dimension',
    'ean',
    'manufacturer',
    'manufacturerNumber',
    'upc',
    'weight'
]

for col in cols:
    print(col)
    print(df[col].unique())
    print(df[df[col].notna()].shape[0])
    print(df.groupby(col)['id'].nunique())
    print('\n')
```

```python
# drop cols

cols_to_drop = [
    'colors',
    'dimension',
    'ean',
    'manufacturer',
    'manufacturerNumber',
    'upc',
    'weight',
    'reviews.userCity',
    'reviews.userProvince',
    'sizes'
]

df = df.drop(columns=cols_to_drop)
df.head()
```

```python
# drop redundant cols
df = df.drop(columns=['asins', 'keys', 'reviews.sourceURLs'])
df.head()
```

```python
# count price difference between amount max and amount min
count = 0
nonzero_diff = 0
for i, row in df.iterrows():
    prices = json.loads(df.iloc[i].prices)
    for price in prices:
        diff = price['amountMax'] - price['amountMin']
        if diff != 0:
            nonzero_diff += 1
        count += 1
print(nonzero_diff)
print(count)
```

```python
df.groupby('id')['prices'].nunique()
```

```python
# create price dataframe
dfp = df.groupby('id')['prices'].first().reset_index()
dfp.head()
```

```python
price_data = []

for i, row in dfp.iterrows():
    pid = row.id
    brand = df[df.id==pid]['brand'].iloc[0]
    categories = df[df.id==pid]['categories'].iloc[0]
    dateAdded = df[df.id==pid]['dateAdded'].iloc[0]
    dateUpdated = df[df.id==pid]['dateUpdated'].iloc[0]
    name = df[df.id==pid]['name'].iloc[0]
    prices = json.loads(dfp.iloc[i].prices)
    for price in prices:
        price_data.append({
            'id': row.id,
            'brand': brand,
            'categories': categories,
            'dateAdded': dateAdded,
            'dateUpdated': dateUpdated,
            'name': name,
            'currency': price.get('currency', pd.NA),
            'amount': price.get('amountMax', pd.NA),
            'date': price.get('dateSeen', pd.NA),
            'isSale': price.get('isSale', pd.NA),
            'merchant': price.get('merchant', pd.NA),
            'shipping': price.get('shipping', pd.NA),
            'condition': price.get('condition', pd.NA),
        })
        
dfprice = pd.DataFrame(price_data)
dfprice.head()
```

```python
dfprice['date'] = [x[0] for x in dfprice['date']]
dfprice.head()
```

```python
dfprice.groupby('id')['currency'].nunique()
```

```python
dfprice.to_csv('prices_clean.csv', index=False)
```
